```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_main = pd.read_csv("FINAL.csv", 
                      usecols = ['Reports', 'VACperc', 'MEDINC', 'PopOv25wBach', 'PopOv25wHS', 'Property'])



# formula: response ~ predictor + predictor + ... 
est = smf.ols(formula='Property ~ Reports + VACperc + MEDINC + PopOv25wBach + PopOv25wHS', data=df_main).fit()
est.summary()


sns.set_context("notebook", font_scale=1.1)
sns.set_style("ticks")

# Create scatterplot of dataframe
sns.lmplot('Reports', # Horizontal axis
           'Property', # Vertical axis
           data=df_main, # Data source
           fit_reg=True, # Don't fix a regression line
           scatter_kws={"marker": "D", # Set marker style
                        "s": 100}) # S marker size

# Set title
plt.title('311 Reports and Property Crime Reports')

# Set x-axis label
plt.xlabel('311 Reports')

# Set y-axis label
plt.ylabel('Property Crimes')

plt.show()
```

```python
df_main = pd.read_csv("FINAL.csv", 
                      usecols = ['Reports', 'VACperc', 'MEDINC', 'PopOv25wBach', 'PopOv25wHS', 'Violent'])



# formula: response ~ predictor + predictor + ... 
est = smf.ols(formula='Violent ~ Reports + VACperc + MEDINC + PopOv25wBach + PopOv25wHS', data=df_main).fit()
est.summary()

sns.set_context("notebook", font_scale=1.1)
sns.set_style("ticks")

# Create scatterplot of dataframe
sns.lmplot('Reports', # Horizontal axis
           'Violent', # Vertical axis
           data=df_main, # Data source
           fit_reg=True, # Don't fix a regression line
           scatter_kws={"marker": "D", # Set marker style
                        "s": 100}) # S marker size

# Set title
plt.title('311 Reports and Violent Crimes Reported')

# Set x-axis label
plt.xlabel('311 Reports')

# Set y-axis label
plt.ylabel('Violent Crimes')

plt.show()
```

```python
import statsmodels.api as sm

df_main = pd.read_csv("FINAL.csv", 
                      usecols = ['Reports', 'Violent'])

X = df_main['Reports']
y = df_main['Violent']

## fit a OLS model with intercept on TV and Radio
X = sm.add_constant(X)
est = sm.OLS(y, X).fit()

est.summary()

```
