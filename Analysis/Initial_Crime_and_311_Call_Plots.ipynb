```python
import pandas as pd
df = pd.read_csv("edited_calls.csv", usecols=['SERVICE REQUEST TYPE', 'Community Area'])

dff = df.groupby('Community Area')['SERVICE REQUEST TYPE'].count().sort_values(ascending = False)
dff = dff.head(15)

ax = dff.plot(kind='bar', figsize=(20,10), title='311 Calls Per Community Area')
ax.set_ylabel("Number of 311 Calls")
```

```python
df = pd.read_csv("edited_crimes2.csv")

df['Burglary'] = df['Primary Type'] == 'BURGLARY'

a = df.groupby(by=['Community Area'])['Burglary'].sum().sort_values(ascending = False) #create table of ward, homicide, and weapons 
a = a.head(15)

ax = a.plot(kind='bar', figsize=(20,10), title='Burglaries Per Community Area')
ax.set_ylabel("Number of Burglaries")
```

```python
df = pd.read_csv("edited_crimes2.csv")

df['Theft'] = df['Primary Type'] == 'THEFT'

a = df.groupby(by=['Community Area'])['Theft'].sum().sort_values(ascending = False) #create table of ward, homicide, and weapons 
a = a.head(15)

ax = a.plot(kind='bar', figsize=(20,10), title='Thefts Per Community Area')
ax.set_ylabel("Number of Thefts")
```
