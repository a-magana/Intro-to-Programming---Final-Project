```python
import pandas as pd
acs = pd.read_csv("ACS_data10.csv",  index_col = "Community Area")
calls = pd.read_csv("311calls.csv",  index_col = "Community Area")
acscalls = acs.join(calls) #joining the acs data with 311 calls and saving as new dataframe, indexed on community area

acscalls.to_csv('acs_calls_merged.csv')
acscalls.head()
```

```python
import pandas as pd
crimesformatted = pd.read_csv("crimes_formatted.csv",  index_col = "Community Area")
acscallz = pd.read_csv("acs_calls_merged.csv",  index_col = "Community Area")
final = crimesformatted.join(acscallz) #joining crime with the merged acs/311 data, indexed on community area

final.head()
final.to_csv('final.csv')
```
